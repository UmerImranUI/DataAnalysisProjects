```python
df2=df.sort_values(by='city')
df2.interpolate()
df2.set_index('date')


```

```python
df3=df2.pivot(index='city',columns='date')
df3
```

```python
new_df=df3.interpolate()
new_df=df3.fillna(method='bfill')
new_df

```

```python
new_df=new_df.fillna(method='ffill')
new_df
```

```python
dfa=df2.pivot_table(index='city',columns='date',aggfunc='sum',margins=True)
dfa
new_df=dfa.interpolate()
new_df=dfa.fillna(method='bfill')
new_df

```
