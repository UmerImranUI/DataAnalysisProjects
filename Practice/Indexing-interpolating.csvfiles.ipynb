```python
import pandas as pd
df=pd.read_csv("F:\py\pandas\weather.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':"no events"
})
```

```python
new_df=df.fillna(method='ffill')  #it fills the next NA to previous oonedata
```

```python
new_df
```

```python
new_df=df.fillna(method='bfill')  #it fills the next NA to next oonedata
```

```python
new_df=df.fillna(method='bfill',axis='columns')  #if copies data horizontally
```
