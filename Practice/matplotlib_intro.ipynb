```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y,color='red',marker='D')
```

```python
plt.plot(x,y,color='red',marker='D',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1 (opacity)
```
